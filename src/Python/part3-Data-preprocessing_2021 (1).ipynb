## 3-2. NumPyオブジェクトの便利なメソッド

## 3-3. NumPyの知ってるといいことあるかもしれない乱数、分布生成メソッド
NumPyは行列だけじゃないんやで！！と言いたくなっちゃうような機能をまとめます。  
特に乱数や分布の話は知っておいて損ないです。たぶん。

````python
# 乱数を生成してみる
print("random:")
rand = np.random.random()
print(rand)
print("- " *30)

print("random_list:")
rand = np.random.random(10)
print(rand)
print("- " *30)

# 整数の乱数
print("random_int:")
rand_int = np.random.randint(5,25,10)
print(rand_int)
print("- " *30)

# 一様分布
print("uniform:")
uniform = np.random.uniform(5)
print(uniform)
print("- " *30)

# 二項分布
print("binomial:")
binomial = np.random.binomial(100,0.5,10)
print(binomial)
print("- " *30)

# 標準正規分布
print("rand_normal:")
rand_normal = np.random.randn(10)
print(rand_normal)
print("- " *30)

# 正規分布(パラメータ指定)
print("normal_dist:")
normal = np.random.normal(60,5,10)
print(normal)
print("- " *30)

# ポアソン分布
print("poisson_dist:")
poisson = np.random.poisson(7,10)
print(poisson)
print("- " *30)

# 指数分布
print("exponential_dist")
exp_dist = np.random.exponential(7,10)
print(exp_dist)
print("- " *30)
````

分布やら乱数やらがいまいち分からんって人は今はなんとなくでもいいかも。でもこの辺りは**大学卒業するまでには押さえておいたほうがいい**と思います。

## 4-1. 拡張子ってなに？
具体的には以下のようなものが拡張子に該当します。
- テキストファイル : .txt .csv  .json など
- Officeファイル : .docx .xlsx .pptx など
- プログラムファイル : .py .c .cpp .rb など
- アプリケーションファイル : .exe .app など
- 圧縮ファイル : .zip .7z など
- 画像ファイル : .jpg .png など
- 動画、音声ファイル : .mp4 .mp3 など

聞いたことがあるものも多いのでは？要はファイルの後ろのアレですね。  
よくある情報系の勉強であるお話に、コンピュータさんは0と1しか分からないとかいう話があります。アレはマジです。なのでそんなコンピュータさんにしっかりと処理をさせたいがために、**ファイルの中身の種類と拡張子の名前とで統一しておきたい**わけです。  
統一しておけば、この拡張子のファイルはこういう書式で書かれているから、このアプリ（ソフト）で開いてあげればうまく処理できますよーとサクサク動ける仕組みです。(だいぶ噛み砕いているので若干違う気がしますがとりあえずイメージはこんな。)  
その証拠にOfficeファイルを右クリックしてメモ帳で開いてみましょう。訳わからん文字が並んでいると思います。これはOfficeのアプリで読み取れるような書式でファイルの中身を記述しているために、我々がメモ帳で読んでみても訳わからんことになってるわけですね。  
（ちなみにOfficeというのは、Word,Excel,PowerPointなどのMicroSoft社のアプリケーションスイートの総称です。それっぽく書いてるから分かりにくいかもだけど、要はWordとかのことですね。）

これができると、ファイルの情報をプログラム中で利用することができます。便利！すごい！...のですが、今やったのはファイルの読み込み(read)です。ここまで来たら書き込み(write)もしたくない？  
てことで読み込みと書き込みの説明のために、先ほどのセルの簡単な解説を。（要所しか解説しません。自分で調べるのだ！）
- open()関数...ファイルを開くよ〜〜っていう関数。
- open()の引数 mode...ファイルを開くときのmodeを指定する。  
    r→read。ファイルを読み込みのモードで開いてね！っていう設定  
    w→write。ファイルを書き込み(上書き)のモードで開いてね！っていう設定  
    a→add。ファイルを書き込み(追記)のモードで開いてね！っていう設定  
    x→x。読み込み、書き込みの区別なくファイルを開く。read, write両方の処理ができるけど、ごちゃごちゃしちゃうので非推奨。  
    b→binary。バイナリモードでファイルを開いてね！っていう設定。まだ難しいので気にしなくてOK!  
- FileObject.readlines()→ファイルの読み込み用メソッド。実は他にも読み込みメソッドがある。調べてみよう！
- FileObject.write()→ファイルの書き込み用メソッド。実は他にも書き込みメソry)
- encoding=""→文字コードのエンコードを指定する。あとでまとめる。

とまあこんなところでしょうか。

ごちゃごちゃしてきましたが、結局のところオブジェクトというのは**モノ**です。ほんでPythonでは**全てのモノがオブジェクト**です。  
整数とか少数とかの数字Objectだし、文字列もObject、リストも辞書も関数もメソッドもぜーんぶがObejectです。  
そしてここは少し難しいので分からなくていいですが、オブジェクトにはclassというものが振られていて、それによって設計されている属性やメソッドを使用することができます。  
例をあげると以下のようなものになります。この詳細はちょっと難易度が高いので触れません。こんなのがPythonにはあるんだ〜〜ぐらいの気持ちでみてくれたらOKです。  
(詳細の資料は2019年度の資料の第5回にのってます。興味があったらみてみてください！)

````python
# さっきからちょいちょい出ているclassというのを新規に定義してみる。
class Student():
    # コンストラクタを定義(イニシャライザとも言う)
    def __init__(self, First_name, Last_name, age):
        # 属性を定義する(attribute)
        self.first_name = First_name
        self.last_name = Last_name
        self.age = age
        self.name = First_name + ", " + Last_name
        self.point = 0
        self.refresh = 0
    
    # メソッドを定義する(method)
    def intoroduce(self, time=12):
        if time < 10 and time > 6:
            word = "Good morning!"
        elif time < 14:
            word = "Hello!"
        elif time < 18:
            word = "Good evening!"
        else:
            word = "Good night!"
        print(word)
        print("My name is " + self.name + ".")
        print("I'm", self.age, "years old.")
    
    def study(self, subject):
        print("I study " + subject + " now...")
        self.point += 1
        
    def sleep(self):
        print("快眠なう。。。")
        self.refresh += 1
        
````

````python
# 上で作ったclassを使ってみる
person = Student("Shigenobu", "Okuma", 30)

# 型は？
print(type(person))
print("- "*30)

# 定義した属性を呼び出してみる
print(person.first_name)  # 属性はカッコいらない
print(person.last_name)
print(person.age)
print(person.name)
print("- "*30)

# 定義したメソッドを呼び出してみる
person.intoroduce()  # メソッドはカッコをつけて呼び出す
print("- "*30)

person.study("English")  # 定義によっては引数が必要

# コメントアウトを外して実行してみてね
#for i in range(5):
#    person.study("English")
#print(teacher.point)
````

これまで説明してきたメソッドの正体に関してもなんとなくイメージできたかもしれません。  
例えばlist。**あれを定義するときどうしてましたか？listの順番を昇順に並び替えるときどうしてましたか？listの要素を参照するときどうしてましたか？要素を追加したいときはどうしてましたか？**  
全てはlistというものがPythonの裏側でlistのclassとして書かれていて、それを我々が呼び出しているに過ぎないのです。  
メソッドはclass内で定義された関数のことを指しています。

### 6-2-1. NumPyのrandomモジュールを利用した分布の生成
上の方の処理を参考にして、以下の分布を生成するプログラムをまとめてみよう。
- 二項分布
- 標準正規分布
- ポアソン分布
- 指数分布
- ベータ分布
- ガンマ分布

とりあえずこれらを10個ずつ表示したlist(numpy.ndarray Object)を作ってみましょう。

````python
# random以下から書いてみよう。引数とかでエラーを吐いたらまずは調べてみよう！
import numpy as np

# 二項分布
print("binomial:")
binomial = np.random.binomial(100,0.5,10)
print(binomial)
print("- " *30)

# 標準正規分布
print("rand_normal:")
rand_normal = np.random.randn(10)
print(rand_normal)
print("- " *30)


# ポアソン分布
print("poisson_dist:")
poisson = np.random.poisson(7,10)
print(poisson)
print("- " *30)


# 指数分布
print("exponential_dist")
exp_dist = np.random.exponential(7,10)
print(exp_dist)
print("- " *30)

# beta分布
print("beta")
beta = np.random.beta(2,10,10)
print(beta)
print("- " *30)

# ガンマ分布
print("gamma")
gamma = np.random.gamma(10,3,10)
print(gamma)
print("- " *30)

random_list = {
    "binomial":binomial,
    "rand_normal": rand_normal,
    "poisson_dist": poisson,
    "exponential_dist":exp_dist,
    "beta": beta,
    "gamma":gamma,


}
````

````python
import matplotlib.pyplot as plt

# jupyterでmatplotを使うときはこの一文を入れることでファイル内で表示してくれる
%matplotlib inline

# DataFrame内の各列に対してforループを回す
for column_name in df.columns:
    plt.figure()  # 新しいプロットを作成

    # ヒストグラムをプロット
    df[column_name].hist(bins=100, color="blue", grid=True, label='pandas')
    plt.ylabel('frequency')
    plt.xlabel('index')
    plt.legend()
    plt.title(f'{column_name}_histgram_norm')

    # 各列ごとに異なるファイルに保存
    plt.savefig(f"{column_name}_hist.png")

    # 表示
    plt.show()
    plt.close()
````
